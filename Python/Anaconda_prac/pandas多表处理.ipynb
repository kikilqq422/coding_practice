```python
import pandas as pd
```

```python
f = pd.read_csv('dataAnalyst1.csv',encoding = 'gbk')
f
```

```python
f.avg.sort_values()
f.sort_values(by = 'bottom',ascending = True)
f.sort_values(by = ['bottom','city'],ascending = True)
```

```python
#将表格转置
f.T
```

```python
#薪资越高，排名越靠前，降序排列,一般method按平均值排名，比如有相同的4个数都在2，则这四个数就是2.5.如果是最大的话，就是这四个数的最后一位排的顺序来算
#first表示不考虑并列的情况
f['rank'] = f.avg.rank(ascending=False, method='min')
f['rank'] = f.avg.rank(ascending=False, method='max')
f['rank'] = f.avg.rank(ascending=False, method='first')
f
```

```python
f.sort_values(by = 'avg',ascending = False)
```

```python
#重复值，重复值计算
f.info()
f.city.unique()
f.city.value_counts()
```

```python
#常见的数据分析维度
f.describe()
f.avg.describe()
f.sum()
f.mean()
f.median()
```

```python
#累计的计算
f['cumsum'] = f.avg.cumsum()
f
```

```python
#分层,分组函数，分层函数,bins是分成几组，labels是分组的名称
f['bins'] = pd.cut(f.avg, bins = 5,labels=list('abcde'))
f
```

```python
f['bins'] = pd.cut(f.avg, bins = [0,10,20,30,100],labels=('0-10','10-20','20-30','30-100'))
f.query('avg >15')
```

```python
#用for循环获取分组后的明细,算出每个城市薪资的最大值和最小值
f.groupby(by = ['city','workYear'])
for k,v in f.groupby(by = ['city']):
    print(k)
    print(v.avg.max()-v.avg.min())
    print(max(v.avg)-min(v.avg))

```
