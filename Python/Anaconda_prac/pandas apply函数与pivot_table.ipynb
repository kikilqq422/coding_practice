```python
import pandas as pd
import numpy as np
```

```python
#avg已经是series了，所以可以切，如果想多维切片求数据，则用loc
pos.groupby(by = ['city','education']).mean().avg['上海']['博士']
pos.groupby(by = ['city','education']).mean().loc['上海','博士']
pos.groupby(by = ['city','education']).mean().reset_index().head()
```

```python
#不用groupby的另外一种方式,直接用set_index是未加工的，先用sort_values进行排序在进行加工
pos.set_index(['city','education'])
pos.sort_values(['city','education']).set_index(['city','education'])
```

```python
#pandas文本修正,将原本的"[]"去掉。去掉引号‘’
pos.positionLables.str[1:-1].str.replace("","")
pos.city.count()
```

```python
#用loc做行索引,行标签才行
pos.loc[980]
pos.loc[pos.city == '深圳','city'] = np.NAN
pos.dropna()
#刪除里面的控制
```

```python
pos = pos.fillna('深圳')
pos.duplicated()
#查找是否有重复值，并且切出重复值,保留重复值的最后一个，或者第一个first
pos[pos.duplicated(keep = 'last')]
pos[~pos.duplicated()]
pos.drop_duplicates().head()
#~返回非重复值，唯一值
```

```python
#apply可以对表进行函数处理,axi=0为函数应用到每个列，axis=1运用到每行
pos.avg.apply(lambda x: str(x)+'k')
def func(x):
    if x > 20: return '20+'
    else: return '< 20'
pos.apply(lambda x :func(x.avg),axis=1)

def func(x):
    if x > 20: return '20+'
    else: return '<20'
pos.avg.apply(func)
#针对每行的x中x.avg进行判断
def func(x):
    if x.avg > 20: return '20+'
    else: return '<20'
pos.apply(func,axis=1)
```

```python
#求每个城市工资排名前5/3的职位明细,用法很灵活
def func(x, n, asc):
    r = x.sort_values('avg',ascending = asc)[:n]
    return r 
pos.groupby(by = 'city').apply(func, n=3, asc = False)
```

```python
#agg函数的用法， 聚合，求和等，apply更加灵活，agg是聚合后的结果
pos.groupby('city').sum()
pos.groupby('city').mean()
pos.groupby('city').agg(['sum','mean'])
pos.groupby('city').agg(lambda x: max(x)-min(x))
```

```python
#数据透视表.分组，分别聚合
pos.pivot_table(
        index = ['city','education'],
        columns = 'workYear',
        values =['avg','top'], 
        aggfunc={'avg':np.mean,
                 'top':np.sum}
        )
```
